```python
import os
import csv


# Function: Analyze Election Data
def analyzeElectionData(input_csvFilePath, input_textFilePath):

    # Declare a dictonary to store election count data {"CaqndidateName" : NumOfVotes}
    electionData = {}
    # Declare a dictonary to store election result data {"CaqndidateName" : PercentOfVotes}
    electionResults = {}
    
    # Read the input CSV (default delimiter = ',') as csvReader
    with open(input_csvFilePath, newline='') as csvFile:
        csvReader = csv.reader(csvFile, delimiter=',')
        # Read header row first (to skip it)
        csvHeader = next(csvReader)
        
        # Read each row of data (csvDataRow) after the header and perform the calculations 
        # Use enumerate to keep track of count  
        # Initialize candidate 
        candidateName = " "
        for csvDataCount, csvDataRow in enumerate(csvReader):
            candidateName = csvDataRow[2]
        
            # Check if candidate exists in electionData dictionary 
            # Add to count if exists 
            if candidateName in electionData.keys():
                electionData[candidateName] = electionData[candidateName] + 1
            # Add entry for the candidate if does not exists
            else:
                electionData[candidateName] = 1

    # Done reading CSV - Print & output the results:
    # Print the result to a text file (passed in as input_textFilePath)
    outputTextFilePath = input_textFilePath
    outputTextResultsFile = open(outputTextFilePath,mode='w')
    # At this point outputTextResultsFile is a pointer to the file 
    outputTextResultsFile.writelines("Election Results" + "\n")
    print("Election Results")
    outputTextResultsFile.writelines("-------------------------" + "\n")
    print("-------------------------")
    print("Total Votes: {}".format(csvDataCount + 1)) 
    outputTextResultsFile.writelines("Total Votes: " + str(csvDataCount + 1) + "\n")
    outputTextResultsFile.writelines("-------------------------" + "\n")
    print("-------------------------")
    # Perform % calculation & Winner
    winner = " "
    winnerPercent = 0.000
    for name, numOfVotes in electionData.items():
        percentOfVotes = round((numOfVotes/(csvDataCount + 1))*100,0)
        # Save the winner
        if ((winnerPercent == 0.000) or (percentOfVotes > winnerPercent)):
            winner = name
            winnerPercent = percentOfVotes
        # print candidate votes & percent
        outputTextResultsFile.writelines(name + ": " + str(format(percentOfVotes,'.3f')) + "% (" + str(numOfVotes) + ")" + "\n")
        print("{} : {:.3f}% ({})".format(name,percentOfVotes,numOfVotes))
        
    outputTextResultsFile.writelines("-------------------------" + "\n")
    print("-------------------------")
    outputTextResultsFile.writelines("Winner: " + winner + "\n")
    print("Winner: {}".format(winner))
    outputTextResultsFile.writelines("-------------------------")
    print("-------------------------")

    

# Main logic:

# Step 1: 
# Build the CSV file path (input)
fileName = "election_data.csv"
csvFilePath = os.path.join('.' , fileName)

# Step 2: 
# Build the text file path (output file)
outputTextFile = "./Election_Analysis_Output.txt"

# Step 2
# Invoke function to analyze election data for the file specified 
analyzeElectionData(csvFilePath, outputTextFile)

```
